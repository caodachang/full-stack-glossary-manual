# TCP 与 UDP

## TCP 是什么？

**TCP**是面向连接的可靠传输协议。

### 三次握手四次挥手

TCP是双向连接，握手时 SYN 包和 ACK 包会一起发，挥手时 FIN 包和 ACK 包大概率是分开的，ACK 的序号是要回复的包的序号加一。

## UDP 是什么？

**UDP**是无连接、不可靠的传输协议。

### 适用场景

实时要求高的场景

## 本篇出现的新名词速查

| 名词 | 一句话解释 |
|------|-----------|
| TCP | 面向连接、可靠的传输协议——三次握手、确认重传、有序交付 |
| UDP | 无连接、不可靠的传输协议——发了不管，极快极简 |
| 端口 | 主机的网络"门牌号"，操作系统根据端口把数据交给对应程序 |
| 三次握手 | TCP 建立连接的过程——SYN→SYN+ACK→ACK |
| 四次挥手 | TCP 关闭连接的过程——双方各发 FIN+响应 ACK |
| ACK | 确认——TCP 每收一个包都告知发送方"收到了" |
| 流量控制 | 接收方根据处理能力限制发送速率 |
| 拥塞控制 | 发送方根据网络拥堵程度调整发送速率 |
| 重传 | 规定时间内没收到 ACK，重新发送该包 |
