# HTTP 概述

## HTTP 是什么？

### Header

万维网三件套：**HTML** 写网页、**URL** 定位资源、**HTTP** 传输数据。

## Cookie 与 Session

HTTP 本身是**无状态**。

**Cookie** 是浏览器端存的一小段数据，服务器通过 `Set-Cookie` Header 让浏览器保存。

**Session**（会话）是服务器端存的状态。Cookie 里只放一个 Session ID，真正的用户数据（登录状态、购物车内容）在服务器内存或 Redis 里。

流程：用户登录 → 服务器创建 Session 存用户信息 → 返回 Set-Cookie 只给 Session ID → 浏览器后续请求自动带 Cookie → 服务器根据 ID 查出 Session。

## 本篇出现的新名词速查

| 名词 | 一句话解释 |
|------|-----------|
| HTTP | 超文本传输协议——浏览器和服务器通信的语言，请求-响应模型 |
| URL | 统一资源定位符——`https://example.com/path?key=value` |
| Cookie | 浏览器端存的小段数据，每次请求自动带，解决 HTTP 无状态 |
| Session | 服务器端存的状态，Cookie 只放 Session ID |
| HTTP/2 多路复用 | 一个 TCP 连接上多个请求帧同时传输 |
| HTTP/3 QUIC | 基于 UDP 的新版 HTTP，单独流的丢包不阻塞其他流 |
