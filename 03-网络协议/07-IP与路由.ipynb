# IP 与路由

## VPN：IP 层的隧道

**VPN**在网络层之上建一条加密隧道。

### 为什么需要 VPN？

*   **加密公网流量**：
*   **访问受限网络**：
*   **跨地域访问**：某些服务只对特定国家开放。VPN 让流量先出国，再从目标国家的服务器出来。

## 本篇出现的新名词速查

| 名词 | 一句话解释 |
|------|-----------|
| IP 地址 | 网络层逻辑地址，标识设备在网络中的位置，软件配置 |
| MAC 地址 | 网卡硬件 ID，出厂写死，链路层用 |
| 子网掩码 | 标记 IP 地址哪段是网络号哪段是主机号——`/24` |
| NAT | 网络地址转换——内网多设备共享一个公网 IP |
| IPv4 | 32 位地址，43 亿个，已分完——NAT 续命中 |
| IPv6 | 128 位地址，多到地球上每粒沙子一个 IP |
| 路由器 | 网络层设备，根据 IP 和路由表跨网络转发数据包 |
| 路由表 | 目标网络→下一跳映射表，路由器根据它决定转发方向 |
| BGP | 自治系统间的路由协议，互联网的"粘合剂" |
| OSPF | 自治系统内的路由协议，自动计算最短路径 |
| TTL | 存活时间——每跳减 1，减到 0 丢弃，防无限环路 |
| ping | 发 ICMP Echo 测连通性和延迟 |
| VPN | 虚拟专用网络——在公网上建加密隧道，让你"出现"在另一个网络 |
| 隧道 | 把完整 IP 包封装进另一个 IP 包，外层加密——VPN 的核心机制 |
