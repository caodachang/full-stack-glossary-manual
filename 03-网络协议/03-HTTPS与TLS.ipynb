# HTTPS 与 TLS

## 历史动机：为什么需要 HTTPS？

在 HTTP 下面加了一层加密，叫 **SSL**。h后标准化为 **TLS**。HTTP + TLS = **HTTPS**

## TLS 是什么？

**TLS** 是夹在 HTTP 和 TCP 之间的一层安全协议。

---

## 本篇出现的新名词速查

| 名词 | 一句话解释 |
|------|-----------|
| HTTPS | HTTP over TLS——加密的 HTTP，防偷看防篡改防冒充 |
| TLS | Transport Layer Security，HTTP 和 TCP 之间的加密层 |
| SSL | TLS 的前身，Netscape 1994 年发布，已废弃但名字还在用 |
| TLS 握手 | 客户端和服务器协商对称密钥的过程 |
