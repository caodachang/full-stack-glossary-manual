# 网络分层模型

## 两种分层模型

### OSI 七层模型

1.  **物理层**：传 0 和 1 的电信号、光脉冲、无线电波。网线、光纤、Wi-Fi 的频段在这一层定义。
2.  **数据链路层**：在两台相邻设备间可靠传输**帧**。交换机在这一层工作，用 **MAC 地址**识别设备。
3.  **网络层**：路由器在这一层工作，用 **IP 地址**定位。这一层的数据单元叫**包**。
4.  **传输层**：在两台主机的进程之间建立通道。**TCP** 提供可靠传输，**UDP** 不确认但快。数据单元叫**段**。
5.  **会话层**：管理对话在这层。实际网络中这层很少作为独立协议出现。
6.  **表示层**：数据格式转换——加密解密、压缩解压、编码转换。
7.  **应用层**：你用的协议。HTTP、DNS、SMTP、WebSocket 都在这里。

### TCP/IP 四层模型

| OSI（七层，理论） | TCP/IP（四层，实际） | 这一层干什么 | 代表 |
|:---:|:---:|---|---|
| 7-5 应用+表示+会话 | 应用层 | 为用户提供服务 | HTTP、DNS、TLS |
| 4 传输层 | 传输层 | 端到端可靠传输 | TCP、UDP |
| 3 网络层 | 网络层 | 跨网络寻址路由 | IP、ICMP |
| 2-1 链路+物理 | 网络接口层 | 相邻设备间传输 | 以太网、Wi-Fi |

## 本篇出现的新名词速查

| 名词 | 一句话解释 |
|------|-----------|
| 网络分层 | 把通信拆成独立层级，每层只做一件事，下层为上层服务 |
| OSI | 七层理论模型——物理→链路→网络→传输→会话→表示→应用 |
| TCP/IP | 四层实际模型——网络接口→网络→传输→应用，互联网的真实骨架 |
| 物理层 | 传 0 和 1 的信号——网线、光纤、Wi-Fi 频段 |
| 数据链路层 | 相邻设备间可靠传输帧，用 MAC 地址识别，交换机工作在这一层 |
| 网络层 | 跨网络路径选择，用 IP 地址定位，路由器工作在这一层 |
| 传输层 | 端到端进程间通信，TCP 可靠，UDP 不确认但快 |
| 应用层 | 面向用户的协议——HTTP、DNS、SMTP、WebSocket |
| 封装 | 数据从上层往下走，每一层加上自己的头部信息 |
| 解封装 | 数据从下层往上来，每一层拆掉对应的头部 |
| MAC 地址 | 网卡出厂时烧录的唯一硬件 ID，链路层用 |
| IP 地址 | 网络层寻址用的逻辑地址，标识主机在网络中的位置 |
| 帧 / 包 / 段 | 不同层的数据单元名字——链路层叫帧，网络层叫包，传输层叫段 |
| 路由器 | 工作在网络层的设备，根据 IP 地址决定数据包走哪条路 |
| 交换机 | 工作在链路层的设备，根据 MAC 地址在局域网内转发帧 |
