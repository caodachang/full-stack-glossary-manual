# DNS

## DNS 是什么？

**DNS** 把域名翻译成 IP 地址。

### CDN：DNS 的直接受益者

**CDN**把静态资源提前缓存到全球各地的**边缘节点**上。用户访问时 DNS 返回离用户最近的节点。

## 本篇出现的新名词速查

| 名词 | 一句话解释 |
|------|-----------|
| DNS | 域名系统——把域名翻译成 IP 地址的全球分布式数据库 |
| 根 DNS | 全球 13 组根服务器，知道所有顶级域服务器的地址 |
| TLD | 顶级域——`.com` `.org` `.cn`，各自维护下一级域名信息 |
| 权威 DNS | 你买域名时配置的 DNS 服务器，知道最终 IP |
| 递归解析器 | 替你从根一路问到权威的跑腿角色 |
| A 记录 | DNS 记录类型——域名→IPv4 地址 |
| CNAME | DNS 记录类型——域名→另一个域名，别名 |
| hosts 文件 | 本机手动维护的域名-IP 映射表，DNS 的前身 |
| 智能 DNS 解析 | 根据请求来源地理位置返回不同 IP——CDN 的核心机制 |
| CDN | 内容分发网络——全球边缘节点缓存静态资源，就近下载 |
| 边缘节点 | CDN 部署在全球各地的缓存服务器，离用户最近 |
| 回源 | CDN 节点缓存没命中时，向源站请求最新内容 |
| Cloudflare | 全球最大免费 CDN，附带 DDoS 防护、DNS、WAF |
