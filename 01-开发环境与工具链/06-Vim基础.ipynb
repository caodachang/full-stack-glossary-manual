# Vim 基础

## Vim 是一个带模式的文本编辑器

```
                      按 i / a / o            
                   输入模式（正常打字）←── 命令模式（敲键盘 = 发命令）
                                                  │
                       按 :                       │
                    底线命令模式 ←─────────────────┘
                    （保存/退出/搜索/替换）     
```

Vim 启动后默认在**命令模式**。按 `i` 进入输入模式，按 `:` 进入底线命令模式。任何时候按 `Esc` 都能回到命令模式。

## 怎么用？
`Esc` 回到普通模式，然后输 `:q!` 回车（不保存退出）或 `:wq` 回车（保存并退出）。

### 打开与退出

````python
vim notes.txt           # 打开（或新建）文件，按 i 进入插入模式编辑

# 先按 Esc 进入命令模式
:q!        # 不保存强制退出
:wq        # 保存并退出（= :w 保存 + :q 退出）
````

### 移动光标

移动键 `h` `j` `k` `l`。

## 典型工作流

git commit 场景：

````python
# git commit 自动打开 Vim

# 1. Vim 打开后你在命令（普通）模式，按 i 进入插入模式
i

# 2. 写提交信息，比如：
#    修复了登录页面的空指针异常

# 3. 按 Esc 回到普通模式
Esc

# 4. 保存并退出
:wq
````

改服务器配置文件的场景：

````python
# 1. SSH 到服务器，用 vim 打开配置文件
ssh user@server
vim /etc/nginx/nginx.conf

# 2. 搜索想改的那一行
/path                   # / 光标跳到底线，搜索单词 "path"

# 3. 进入插入模式修改
i                       # 修改……

# 4. 按 n 跳到下一个匹配处再修改

# 5. 保存退出
Esc
:wq
````
