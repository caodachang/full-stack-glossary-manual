# Git 概述与基础操作

## 历史动机：为什么需要 Git？

- **分布式**：每个人本地都有一份完整的仓库。不依赖中央服务器，离线也能工作
- **快**：绝大多数操作在本地完成，速度快
- **分支廉价**：创建分支几乎零成本
- **数据完整**：每个文件、每次提交都用 SHA-1 哈希值唯一标识，历史不可篡改

> Git 跟踪的是**文件的快照**。

## Git 是什么？

**Git** 是一个**分布式版本控制系统**（Version Control System，简称 VCS）。

```
工作区（你写代码的地方）
    ↓ git add（拍一张快照，放到暂存区）
暂存区（你挑选好、准备下一次提交的快照集合）
    ↓ git commit（把暂存区的所有快照正式存档）
本地仓库（.git文件夹，所有快照按时间线排好，可随时查阅）
```

## 怎么用？

安装git后需要告诉 Git 你是谁（每次提交都会带上这个身份）：

````python
git config --global user.name "你的名字"
git config --global user.email "你的邮箱"

# git config user.name 查看名字
# git config user.email 查看邮箱
````

> `--global` 表示全局配置（对你电脑上所有 Git 仓库生效）。Git 的配置分三个层级：`--local`（只影响当前仓库）、`--global`（影响当前用户所有仓库）、`--system`（影响整台机器所有用户）。

### git init：从零开始一个仓库

````python
# 进入项目文件夹后，创建为git仓库
git init
````

### git add：把修改加入暂存区

先创建一个文件来演示：

````python
echo "# 我的项目" > README.md    # 创建一个 README.md 文件
git status                       # 查看当前仓库状态
````

`git status` 告诉你**谁变了、谁在暂存区、谁还没被跟踪**。你会看到 `README.md` 显示为红色 "Untracked files"（未跟踪：Git 还没开始管理这个文件）。

现在把这个文件加入暂存区：

````python
git add README.md       # 把 README.md 加入暂存区
# 或者
git add .               # 把当前目录下所有改动的文件都加入暂存区
````

再次 `git status`，README.md 变绿了——它现在在暂存区，等着被提交。

### git commit：正式存档

````python
git commit -m "初次提交，创建 README"   # -m 后面跟提交信息
````

`git commit` 把暂存区里的所有快照打包，生成一个**提交对象**（commit object），这个对象包含：

- 一个独一无二的 SHA-1 哈希值作为 ID
- 提交时间、作者
- 你写的提交信息（commit message：用一句简短的话描述你这次做了什么）
- 指向这次提交的文件快照
- 指向上一次提交的指针（pointer：存的是上一件事的地址，顺着能往回找——这就是历史链的形成方式）

````python
git commit -m "修复了登录页面的bug"    # 好的提交信息：清楚说明改了什么
git commit -m "更新"                   # 不好的提交信息：太模糊
````

> 提交信息怎么写？约定俗成用英文现在时，如 `add login feature`、`fix button style`、`update docs`。首字母小写，不加句号。

`-m` 是 message 的缩写。不加 `-m` 的话，Git 会打开 Vim 编辑器让你写信息。

### git log：翻阅历史

````python
git log                     # 完整日志
git log --oneline           # 一行一条，精简显示
git log --oneline --graph   # 加上分支图
````

````python
git log --author="你的名字"    # 只看某人的提交
git log -- README.md           # 只看某文件的提交历史
git log -3                     # 只看最近 3 条
````

### git diff：看改了哪里

````python
git diff                  # 工作区 vs 暂存区（还没 add 的修改）
git diff --staged          # 暂存区 vs 上次提交（已经 add 了但还没 commit）
git diff HEAD              # 工作区 vs 上次提交（不管有没有 add，所有改动）
````

`git diff` 的输出中，`-` 号开头表示删掉的行，`+` 号开头表示新增的行。

### git reset：撤销操作

在 Git 中撤销分不同场景：

````python
# 场景1：把某个文件从暂存区撤回来（保留修改，只是不提交它）
git reset HEAD -- README.md

# 场景2：撤销最近一次 commit，修改回到暂存区（提交白做了，改动还在）
git reset --soft HEAD~1

# 场景3：撤销最近一次 commit 并清空暂存区（改动回到工作区）
git reset --mixed HEAD~1

# 场景4：彻底丢弃所有修改回到上次 commit 的状态
git reset --hard HEAD
````

`HEAD~1` 表示「上一次提交」，`HEAD~2` 就是上上次。

## 典型工作流

一个修改从产生到存档的完整流程：

````python
# 1. 写了一堆代码
git status                         # 先看看改了哪些文件

# 2. 挑选要提交的文件
git add src/main.py                # 逐个加入
git add docs/                      # 或者整个目录

# 3. 再确认一下
git status                         # 绿的就是待提交的

# 4. 存档
git commit -m "add main function"

# 5. 看看历史
git log --oneline -5               # 最近5条
````

## 本篇出现的新名词速查

| 名词 | 一句话解释 |
|------|-----------|
| 版本控制 (VCS) | 记录文件每次变化、随时可回退历史的系统 |
| 仓库 (Repository) | 存项目所有文件和历史的地方 |
| 分布式 (Distributed) | 每个人本地都有完整仓库，不依赖中央服务器 |
| 哈希 (Hash) | 把任意数据压缩成固定长度指纹，内容变一点指纹全变 |
| 工作区 (Working Directory) | 你电脑里看到的项目文件夹 |
| 暂存区 (Staging Area) | add 后、commit 前的等待区，精确控制哪些修改进存档 |
| 本地仓库 (Local Repo) | `.git/` 文件夹，存着所有版本历史和快照 |
| 提交 (Commit) | 把暂存区内容正式录入历史的动作 |
| 快照 (Snapshot) | 文件在某瞬间的完整副本，Git 存的是这个，不是差异 |
| HEAD | 指向当前所在提交的指针 |
| 终端 (Terminal) | 敲命令操作电脑的界面 |
| 克隆 (Clone) | 把远程仓库完整复制到本地 |
