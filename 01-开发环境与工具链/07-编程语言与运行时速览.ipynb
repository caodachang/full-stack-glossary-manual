# 编程语言与运行时速览

## 编译型 vs 解释型 vs JIT

**编译型**：（C、C++、Rust、Go ）用编译器把全部源码一次性翻译成机器码，生成可执行文件。

**解释型**：（Python、JavaScript、Ruby）直接用解释器跑，逐行读、逐行翻译、逐行执行。

**JIT（即时编译）**：（Java 的 JVM、JavaScript 的 V8 引擎、Python 的 PyPy ）运行时监视哪些代码被反复执行，发现后把那部分编译成机器码，下次直接跑编译版。

## 静态类型 vs 动态类型

**静态类型**：（C、Java、Rust、Go、TypeScript）变量出生时绑定类型。

**动态类型**：（Python、JavaScript、Ruby ）变量只是名字标签，运行时自动确定类型

## GC：内存管理

**手动管理**：程序员自己申请和释放内存。

**引用计数**：每个对象记着"多少人正在用我"，计数归零就回收。

**GC（垃圾回收）**：运行时定期扫描内存，找出"再也访问不到"的对象回收。

## 主流语言一句话定位

**C** 是语言世界的基石。操作系统、数据库内核、Python 解释器本身都用 C 写。

**C++** 是 C 加上面向对象和泛型。

**Java** 跨平台强
**Python** 优先可读性。

**JavaScript** 是浏览器的唯一母语。

**TypeScript** 是 JavaScript 加上静态类型。

**Go** 是用接近 C 的性能写出接近 Python 的开发体验。

**Rust** 主打内存安全不牺牲性能——编译器极严格，阻止你写出可能崩溃的代码。

## 本篇出现的新名词速查

| 名词 | 一句话解释 |
|------|-----------|
| 编译型 | 源码一次全量编译成机器码再运行（C/Rust/Go） |
| 解释型 | 源码由解释器逐行翻译执行（Python/JS/Ruby） |
| JIT | 即时编译——运行时把热点代码动态编译成机器码 |
| 字节码 | 介于源码和机器码之间的中间格式，由虚拟机执行 |
| 虚拟机 | 在 OS 上模拟出一台计算机，跑字节码（JVM/.NET CLR） |
| 静态类型 | 变量出声时类型定死，编译时检查 |
| 动态类型 | 变量只是标签，运行时才定类型 |
| 内存泄漏 | 不再用的内存没回收，越积越多最终崩掉 |
| 引用计数 | 每个对象计引用次数，归零回收（CPython） |
| GC | 垃圾回收——运行时自动扫描回收不可达对象 |
