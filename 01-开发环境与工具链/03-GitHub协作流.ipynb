# GitHub 协作流

## 历史动机：为什么需要 GitHub？

GitHub：**给 Git 仓库加上一层网页界面和社交协作层**。你在浏览器里就能：

- 托管代码（push 上去就上线，不用自己搭服务器）
- 提 Issue 追踪任务和 bug
- 发 Pull Request 发起代码审查
- 用 Actions 自动测试和部署
- Fork 别人的项目，改完再贡献回去

## 怎么用？

.gitignore：告诉 Git 别管哪些文件

### git remote：关联本地仓库和 GitHub

你在本地用 `git init` 建好仓库、写了几次提交后，怎么把它搬到 GitHub 上？

第一步是在 GitHub 网页上创建一个空仓库（New Repository），拿到它的 URL。

第二步用 `git remote` 把本地仓库和这个远程仓库关联起来：

````python
git remote add origin https://github.com/用户名/项目名.git   # 关联远程仓库，起别名叫 origin
git remote -v                                                # 查看已关联的远程仓库及地址
git remote show origin                                       # 查看 origin 的详细信息

# 删除错误的远端仓库：
git remote remove origin
````

### git push：把本地提交推到 GitHub

````python
git push -u origin main      # 首次推送：把 main 分支推上去，同时绑定上游
git push                     # 绑定后直接 push，不用每次都写 origin main
git push origin feature-x    # 把其他分支也推上去
````

### Fork：把别人的项目复制到自己名下

Fork 是 GitHub 上最核心的协作动词之一。

你想给一个开源项目贡献代码，但你肯定没有直接往别人仓库 push 的权限。Fork 就是解决方案：**把别人的仓库复制一份到你自己的 GitHub 账号下，你对这份副本有完全控制权。**

```
原始仓库（原作者）                你的 Fork（你名下）
github.com/作者/项目               github.com/你的用户名/项目
    ↑                                      ↑
    └── 你没有 push 权限                    └── 你有完全控制权
```

Fork 之后的标准操作流程：

````python
# 1. 在 GitHub 网页上点 Fork 按钮
# 2. 把自己名下的 Fork 克隆到本地
git clone https://github.com/你的用户名/项目.git
cd 项目

# 3. 把原仓库加为第二个远程（取名 upstream），方便同步
git remote add upstream https://github.com/原作者/项目.git
git remote -v        # 查看远程仓库列表

# 4. 同步原仓库的更新到本地
git fetch upstream   # 下载原仓库的更新（不自动合并）
git merge upstream/main  # 把原仓库的 main 分支合并到本地的 main 分支
git push origin main  # 把本地的 main 分支推送到自己名下的 Fork 仓库
````

### Pull Request（PR）：请求别人拉取你的代码

Fork 让你能改代码，但改完后怎么贡献回原项目？需要**Pull Request**（拉取请求，简称 PR）。

PR 的标准工作流：

```bash
# 同步原仓库最新代码
git fetch upstream
git switch main
git merge upstream/main                     # 让本地 main 跟上原仓库

# 创建功能分支
git switch -c fix-typo

# 改代码并提交
git add . && git commit -m "fix typo in README"

# 推到你自己的 Fork
git push -u origin fix-typo
```

然后在 GitHub 网页上：

1. 打开你的 Fork 页面，GitHub 会提示「fix-typo 分支有新提交，要不要开 PR」
2. 点「Compare & pull request」
3. 写标题和描述，说明你改了什么、为什么改
4. 点「Create pull request」

PR 创建后，原作者会看到你的代码变更（diff），可以逐行评论（Code Review，即代码审查：团队成员逐行检查你的代码，指出问题或建议修改）、要求修改、讨论方案。讨论完后原作者点「Merge pull request」——你的代码就合进了原项目。

### Issue：追踪任务和讨论

**Issue**（议题）是 GitHub 上用来追踪工作项的工具——可以是 bug 报告、功能请求、文档改进、问题咨询。修复了会自动关闭

### GitHub Actions：自动化流水线

你每次 push 代码后，可能要做：运行测试、检查代码风格、构建、部署到服务器……手工做又慢又容易漏。

**GitHub Actions** 就是「当某个事件发生时，自动执行你定义的任务」。

> CI/CD 是什么？CI（Continuous Integration，持续集成）指每次改代码后自动跑测试和检查，确保改动没搞坏现有功能。CD（Continuous Delivery/Deployment，持续交付/部署）指测试通过后自动发布到服务器。两者合称 CI/CD Pipeline（流水线：从代码提交到上线的自动化链路），GitHub Actions 就是实现 CI/CD 的一种工具。

## 一句话总结

**GitHub = Git + 社交协作层。** `.gitignore` 过滤垃圾文件，`remote` 关联远程仓库，`push`/`pull`/`fetch` 同步代码，`fork` 把别人项目搬到自己名下，`clone` 下载到本地，`PR` 发起合并请求+代码审查，`Issue` 追踪任务和讨论，`Actions` 自动测试和部署。这九个概念就是 GitHub 协作的骨架。
