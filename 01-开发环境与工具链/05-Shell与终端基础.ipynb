# Shell 与终端基础

## Shell 是在终端里跑的程序

**sh** 最简单，最兼容，交互能力极弱，主要用来写安装脚本

**bash** 加了交互能力，服务器上一定有它，可用来写服务器运维脚本

**zsh** 兼容 bash ，配制 oh-my-zsh 后有更好的交互能力

**fish** 不兼容 bash，无需配置开箱即用

**CMD/PowerShell** 微软的 Shell，打开终端后＋⬇可以选择其他程序

## 使用方法

````python
pwd                    # 打印当前路径
ls                     # 列出当前目录下的文件和文件夹
cd Documents/          # 进入 Documents 文件夹


touch notes.txt             # 创建文件
mkdir my-project            # 创建文件夹
mkdir -p a/b/c              # 嵌套创建文件夹
rm notes.txt                # 删除文件

cp notes.txt backup.txt        # 复制 notes.txt → backup.txt
mv old-name.txt new-name.txt   # 重命名

cat notes.txt                  # 查看文件
````

## PATH 与环境变量

**环境变量** 有用户主目录，当前登陆用户名……等内容，所有程序都能读取。PATH 就是其中一个指定 Shell 去哪找命令。

### 重定向与管道

**重定向**把数据流从默认的键盘/屏幕改到文件：

*   `>` ：把标准输出（报错不会）写到文件
*   `>>`：追加到文件末尾
*   `<` ：从文件读取当输入

**管道** `|` 把左边命令的标准输出直接变成右边命令的输入

### 文件权限

ls 后标准输出

```
-rw-r--r--  1 user group  4096 Jun 27 10:00 notes.txt
 ↑  ↑  ↑
 │  │  └── others（其他人）权限
 │  └───── group（同组）权限
 └──────── owner（文件主人）权限

r = read（可读）   w = write（可写）   x = execute（可执行）
```

## 典型工作流

````python
# 1. 看看自己在哪
pwd                                    # /Users/yourname

# 2. 跳到项目目录
cd ~/projects

# 3. 创建项目文件夹（&& 表示前一条成功才执行后一条）
mkdir my-web-app && cd my-web-app

# 4. 创建初始文件
touch index.html style.css app.js
mkdir images scripts

# 5. 确认结果
ls -la

# 快捷键 Ctrl + l 清屏
````

## 本篇出现的新名词速查

| 名词 | 一句话解释 |
|------|-----------|
| 命令行界面 (CLI) | 通过文本命令与计算机交互的界面 |
| 图形化界面 (GUI) | 通过窗口、按钮、图标与计算机交互的界面 |
| Shell | 命令解释器，在终端里跑的程序 |
| 终端 (Terminal) | 容纳 Shell 的窗口程序，负责显示和接收输入 |
| 脚本 (Script) | 一系列命令写成文件，可一次性自动执行 |
| 环境变量 | Shell 内存中的键值对，所有程序都能读取 |
| PATH | 特殊环境变量，存着 Shell 查找可执行程序的目录列表 |
| 标准输入 (stdin) | 程序读取输入数据的数据流 |
| 标准输出 (stdout) | 程序正常输出数据的数据流 |
| 标准错误 (stderr) | 程序报错时独立的输出数据流 |
| 重定向 | 把命令的输入/输出从默认的键盘/屏幕改为文件 |
| 管道 (Pipe) | `\|` 符号，左边命令的 stdout 直接变成右边命令的 stdin |
| bash | 最通用的 Shell，服务器标配 |
| zsh | bash 的增强版，macOS 默认 Shell |
| WSL | 在 Windows 里跑一个真正的 Linux |
