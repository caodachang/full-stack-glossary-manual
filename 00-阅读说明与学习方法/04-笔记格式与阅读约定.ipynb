# 笔记格式与阅读约定

## 先说结论

本手册使用 .ipynb 作为笔记容器，但它不是代码实验 notebook。默认所有内容都写在 markdown cell 中，代码只作为展示示例，不要求运行。

这样做的目的很简单：让读者专注于理解概念和规格表达，而不是被 kernel、运行环境、依赖安装、代码报错打断。

## 为什么使用 .ipynb

.ipynb 有几个适合学习手册的优点：

- 可以分 cell 组织内容，阅读节奏清楚
- markdown、表格、代码块显示效果较好
- 以后如果某些章节真的需要演示，可以单独加入可运行 cell
- 在 VS Code、Cursor、Jupyter 等工具里都比较容易打开

但本手册目前不把“运行代码”作为目标。

## 默认只使用 markdown cell

每篇笔记默认由多个 markdown cell 组成。

推荐拆分方式：

~~~text
cell 1：标题和先说结论
cell 2：为什么需要它
cell 3：它是什么
cell 4：它不是什么
cell 5：相邻概念对比
cell 6：怎么使用或怎么表达
cell 7：在规格里怎么写
cell 8：常见误区和面试扩展
cell 9：延伸阅读方向
~~~

这样读者可以一段一段读，也方便后续修改。

## 代码示例怎么写

所有代码、配置、命令、接口示例，都写在 markdown fenced code block 中。

例如：

~~~markdown
~~~http
GET /orders?page=1&page_size=20
Authorization: Bearer <token>
~~~
~~~

或者：

~~~markdown
~~~json
{
  "id": "order_001",
  "status": "paid"
}
~~~
~~~

这些代码块只用于展示结构，不要求执行。

## 为什么不默认使用 code cell

code cell 容易带来几个问题：

- 读者可能误点运行，出现无意义报错
- 一个 notebook 通常只有一个 kernel，但示例可能涉及 bash、json、yaml、http、python、javascript 多种语言
- code cell 的语言 metadata 可能影响高亮和警告
- 本手册重点不是运行，而是理解和表达

所以默认不用可执行 code cell。只有未来某篇明确需要实验演示时，才单独说明。

## 单篇笔记的推荐结构

每篇笔记建议采用这个结构：

~~~text
# 名词标题

## 先说结论
3-5 句话告诉读者它是什么、解决什么问题。

## 为什么需要它
从真实场景或历史动机讲起。

## 它是什么
给出清晰定义，少堆术语。

## 它不是什么
排除常见误解。

## 和相邻概念的区别
横向对比，说明边界。

## 怎么使用
给最小可理解示例。

## 在规格里怎么写
说明它是否应该出现在软件规格中，以及写到什么粒度。

## 面试常见问法
如果它是高频面试点，适度扩展。

## 常见坑
列出初学者容易踩的坑。
~~~

## 阅读深度约定

不是所有概念都需要同样深度。

### 必须精读

和写规格直接相关的内容，例如需求、验收标准、API、数据、权限、错误状态、测试。

### 建立位置感即可

一些底层或规模化概念，例如 io_uring、Kubernetes、共识算法、GPU/CUDA。第一次读只需要知道它们解决什么问题，什么时候会出现。

### 用到再深挖

具体框架 API、平台配置细节、云服务操作步骤、模型训练参数。这些内容变化快，最好在实际项目中结合官方文档学习。

## 面试扩展的写法

如果某个概念是面试高频点，可以增加一节：

~~~text
## 面试常见问法

- 常见问法：这个概念通常怎么问？
- 考察重点：面试官真正想确认什么？
- 回答边界：初学者回答到什么深度就够？
- 容易误区：哪些回答像背八股但没有理解？
~~~

这部分只作为理解加深层，不把笔记变成题库。

## 文件命名约定

文件名要让读者一眼知道主题。

推荐：

~~~text
01-需求、PRD、Spec与SDD.ipynb
09-REST：理解常见API设计风格.ipynb
12-OpenAPI与Swagger：把接口写成协作文档.ipynb
~~~

不推荐：

~~~text
01-basic.ipynb
02-advanced.ipynb
03-note.ipynb
~~~

编号用于控制阅读顺序，标题用于表达内容边界。
