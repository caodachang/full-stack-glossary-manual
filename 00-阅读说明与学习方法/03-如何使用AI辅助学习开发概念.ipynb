# 如何使用 AI 辅助学习开发概念

## 先说结论

AI 很适合帮你解释概念、生成例子、做横向对比、检查规格遗漏。但 AI 不应该替你判断所有东西是否正确。

学习开发概念时，最好的方式不是问“某某是什么”，而是让 AI 把概念放回场景里解释，并要求它说明边界、误区、规格写法和验收标准。

## 不推荐的问法

很多人会这样问：

~~~text
REST 是什么？
JWT 是什么？
Docker 是什么？
~~~

这种问法能得到定义，但很容易停在百科层面。你会知道一个词的大概意思，却不知道它什么时候出现、和其他词有什么区别、写项目时要不要处理。

## 更推荐的问法

可以把问题改成：

~~~text
请用一个真实软件功能解释 REST 为什么会出现。
顺便对比 REST、GraphQL、gRPC。
最后告诉我，如果我要写一个 spec，哪些内容需要写进去。
~~~

或者：

~~~text
我正在设计一个登录功能。
请解释认证、授权、Session、Token、JWT、OAuth 分别在这个功能里处于什么位置。
哪些必须写进早期 spec，哪些可以暂时不写？
~~~

这种问法会迫使 AI 从“定义模式”进入“工程解释模式”。

## 用 AI 学概念的五步法

### 第一步：让 AI 讲场景

~~~text
不要先给定义。请先讲一个我为什么会遇到这个概念的真实场景。
~~~

### 第二步：让 AI 给定位

~~~text
这个概念属于需求、前端、后端、数据、部署、安全、AI 编程，还是系统设计？它在软件流程中什么时候出现？
~~~

### 第三步：让 AI 横向对比

~~~text
请对比它和 3 个相邻概念，说明什么时候用它，什么时候不用它。
~~~

### 第四步：让 AI 转成规格

~~~text
如果我要把这个概念写进 spec，应该写哪些字段、规则、接口、错误状态或验收标准？
~~~

### 第五步：让 AI 找误区

~~~text
初学者最容易误解这个概念的地方有哪些？面试中如果问它，真正想确认什么？
~~~

## 用 AI 检查 spec

当你写完一段规格后，可以让 AI 做检查，而不是让它直接替你写。

示例：

~~~text
请检查下面这段 spec 是否缺少页面状态、接口约定、数据字段、权限、错误处理或验收标准。
不要直接改写，先列出问题和原因。
~~~

这样可以避免 AI 直接生成一大段看似完整但不一定符合你意图的内容。

## 用 AI 生成面试扩展

如果某个概念是面试高频点，可以这样问：

~~~text
这个概念如果出现在面试里，常见问法有哪些？
面试官真正想确认什么？
请给出适合初学者的回答边界，不要写成背诵八股。
~~~

面试扩展的目的不是刷题，而是帮助你确认自己是否真的理解了概念的边界。

## AI 回答需要警惕什么

### 1. 术语堆叠

AI 可能用更多术语解释一个术语。遇到这种情况，要求它用生活类比或真实功能重新解释。

### 2. 过度设计

AI 很容易把简单需求扩展成复杂架构。你要追问：这是当前必须做的吗，还是未来规模变大后才需要？

### 3. 版本和生态变化

工具、框架、平台政策会变化。如果涉及最新版本、价格、政策、平台审核规则，需要查官方资料，而不是只相信模型记忆。

### 4. 生成内容看起来完整但不可验收

没有验收标准的回答都要继续追问：怎样算完成？怎样验证？失败时怎么处理？

## 推荐提示词模板

~~~text
请用“从名词到 spec”的方式解释【概念名】：

1. 先说它为什么会出现
2. 再说它解决什么问题
3. 放到软件开发流程中定位
4. 对比 2-3 个相邻概念
5. 给一个最小真实场景
6. 说明写 spec 时要不要提到它，怎么写
7. 列出初学者常见误区
8. 如果它是面试高频点，补充常见问法和回答边界
~~~

这个模板可以用于本手册后续大多数概念。
