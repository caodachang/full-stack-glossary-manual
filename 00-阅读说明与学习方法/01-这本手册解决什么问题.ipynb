# 这本手册解决什么问题

## 先说结论

这本手册解决的不是“怎么背更多技术名词”，而是“怎么把模糊想法变成人能评审、AI 能执行、结果能验收的软件规格”。

很多人使用 AI 编程工具时，并不是卡在不会写某一行代码，而是卡在不知道该怎么描述一个软件：页面要有哪些状态，接口要提供什么能力，数据要保存什么，权限怎么划分，错误怎么处理，怎样才算完成。

所以这本手册的核心任务，是把软件开发里反复出现的名词放回真实流程中解释。读者不一定要亲手实现所有技术，但需要知道它们为什么存在、什么时候出现、和规格有什么关系。

## 它不是传统代码教程

传统代码教程通常会从一门语言或一个框架开始：变量、函数、组件、接口、数据库、部署。这个路径适合训练写代码，但不一定适合 AI 协作开发时代的初学者。

AI 已经能帮人生成很多代码。真正难的部分变成了：

- 你能不能说清楚要做什么？
- 你能不能判断 AI 生成的方案有没有偏题？
- 你能不能指出接口、数据、权限、错误处理哪里不完整？
- 你能不能写出可验收的完成标准？

这本手册更关心这些问题，而不是让读者从第一天开始背 API 或刷框架用法。

## 它也不是百科词典

百科词典会告诉你 REST 是什么、JWT 是什么、CI/CD 是什么，但往往不会告诉你：为什么你在写一个登录功能时会同时遇到 HTTP、Session、Token、数据库、权限、测试、部署这些词。

本手册按软件从想法到交付的过程组织内容。每个名词都要回答三个位置问题：

1. 它在哪个阶段出现？
2. 它解决什么问题？
3. 写规格时要不要提到它？

也就是说，本手册不是把名词堆起来，而是帮读者建立一张开发流程地图。

## 为什么 AI 编程时代更需要这本手册

使用 AI 编程工具时，输入越模糊，输出越容易看起来很完整、实际很危险。

例如你说：

~~~text
帮我做一个登录功能。
~~~

AI 可能会生成页面、表单、接口、数据库字段、密码校验、Token 逻辑。但如果你自己不理解认证、授权、Session、JWT、密码哈希、错误状态和安全边界，就很难判断生成结果是否可靠。

更好的输入应该接近规格：

~~~text
做一个邮箱密码登录功能。
用户输入邮箱和密码。
密码必须用安全哈希保存。
登录成功后返回会话凭证。
连续失败 5 次需要限制尝试。
需要处理邮箱不存在、密码错误、账号被禁用三种错误状态。
完成后提供页面、接口、数据字段和验收标准。
~~~

这本手册要训练的，就是从第一种表达走向第二种表达。

## 本手册的最终能力目标

读者读完后，不一定能独立手写复杂系统，但应该能做到：

- 听懂常见开发名词，不被术语打断思考
- 知道一个软件从想法到交付大致经过哪些环节
- 能把需求拆成页面、接口、数据、权限、错误处理和验收标准
- 能判断一个概念是当前必须处理，还是未来规模变大后再考虑
- 能和 AI 工具进行更稳定的协作
- 能写出人能评审、AI 能执行、结果能验收的软件规格

## 学习时应该避免什么

### 不要急着把每个概念学到专家级

第一次读到 Kubernetes、OAuth、RAG、分布式事务时，只需要知道它们大概解决什么问题、在哪个阶段出现。等真正需要时，再深入细节。

### 不要把代码块当成必须运行的实验

本手册中的代码和配置示例主要用于展示结构，不要求执行。重点是理解表达方式，而不是把每段代码跑通。

### 不要把面试扩展当成刷题入口

如果某个概念也是面试高频点，笔记会适度补充常见问法。但面试内容是理解加深层，不是主线。主线仍然是建立软件开发认知和规格表达能力。

## 这本手册最适合怎么用

建议把它当作三种工具：

1. **地图**：不知道一个词在哪个环节时，用它定位。
2. **翻译器**：AI 或工程师说出一串术语时，用它翻译成可理解的开发动作。
3. **规格训练器**：写需求时，用它提醒自己是否漏掉页面、接口、数据、权限、测试和验收。

如果你每读完一个名词，都能更清楚地写出一句规格，这本手册就在发挥作用。
