# 如何从名词读到 spec

## 先说结论

学习开发名词的最终目的，不是把定义背下来，而是能把它们转化成软件规格里的清晰约束。

一个名词只有进入规格，才真正参与了软件交付。例如你知道“分页”这个词，还不够；你需要能写清楚列表接口使用 offset 还是 cursor，每页多少条，空结果怎么显示，排序规则是什么，前端如何加载下一页。

这就是“从名词读到 spec”的意思。

## 什么是 spec

这里的 spec 指软件规格说明。它不是一种固定格式，而是一种清晰表达：让人能评审，让 AI 能执行，让结果能验收。

一个好的 spec 通常包含：

~~~text
目标：要解决什么问题
用户：谁会使用
范围：做什么，不做什么
流程：用户如何完成任务
页面：有哪些界面和状态
接口：系统需要提供哪些能力
数据：需要保存哪些对象和字段
权限：谁能看、谁能改、谁不能做
错误：失败时如何提示和处理
验收：怎样算完成
~~~

不是所有项目都需要写得很重，但这些维度要在脑子里有位置。

## 名词到 spec 的转换公式

读一个名词时，可以用这个公式：

~~~text
名词
  -> 它解决什么问题
  -> 它出现在软件流程的哪个位置
  -> 它会影响哪些设计决定
  -> 它在规格里应该怎么表达
~~~

例如“认证”：

~~~text
认证
  -> 解决“你是谁”的问题
  -> 出现在登录、会话、权限之前
  -> 影响登录方式、凭证保存、错误处理、安全策略
  -> 规格中要写登录方式、凭证机制、失败提示、会话过期规则
~~~

## 示例：REST 如何进入 spec

REST 理解 API 规格的核心名词。

如果需求是“用户可以查看自己的订单”，一个模糊说法是：

~~~text
做一个订单列表接口。
~~~

读懂 REST 之后，可以写得更像规格：

~~~text
接口：GET /orders
用途：获取当前登录用户的订单列表
认证：需要登录
查询参数：page、page_size、status
返回：订单 id、标题、金额、状态、创建时间
空状态：没有订单时返回空列表，前端显示空状态提示
错误：未登录返回 401
~~~

这里 REST 的价值不是背定义，而是帮助你把“接口”写得稳定、可讨论、可验收。

## 示例：数据库如何进入 spec

如果需求是“用户可以收藏文章”，只写功能是不够的。

读懂数据建模后，可以补充：

~~~text
数据对象：收藏记录
字段：id、user_id、article_id、created_at
约束：同一个用户不能重复收藏同一篇文章
删除：用户取消收藏时删除收藏记录
查询：文章详情需要返回当前用户是否已收藏
~~~

这不是要求读者立刻设计完整数据库，而是让读者意识到：收藏不是按钮效果，它背后有数据关系和约束。

## 示例：测试如何进入 spec

很多需求失败，不是因为功能没写，而是因为没有定义怎样算完成。

例如“支持搜索文章”，可以补成：

~~~text
验收标准：
- 输入关键词后显示匹配文章列表
- 没有结果时显示空状态
- 搜索请求失败时显示错误提示
- 清空关键词后恢复默认列表
- 搜索结果按相关性或时间排序，排序规则需要明确
~~~

测试相关名词的价值，就在于提醒你把结果变成可验证的条件。

## 读每篇笔记时问自己七个问题

1. 这个词解决什么问题？
2. 它属于需求、前端、后端、数据、部署、安全、AI，还是系统设计？
3. 它和相邻概念有什么区别？
4. 如果不处理它，会出现什么风险？
5. 当前项目早期需要它吗，还是规模变大后再考虑？
6. 它在 spec 里应该写成字段、接口、流程、权限、错误，还是验收标准？
7. 面试中问它时，真正想考的是定义，还是工程判断？

## 常见误区

### 误区一：会背定义就算懂

能背“REST 是一种架构风格”不等于能写接口规格。真正有用的是知道资源怎么命名、方法怎么选、错误怎么返回。

### 误区二：所有名词都要写进 spec

不是。比如 Kubernetes 很重要，但一个早期本地工具项目可能完全不需要在 spec 里写 Kubernetes。规格应该服务当前交付，不是展示知识储备。

### 误区三：spec 写得越长越好

也不是。好的规格是清楚，不是冗长。能让人评审、AI 执行、结果验收，就够了。
