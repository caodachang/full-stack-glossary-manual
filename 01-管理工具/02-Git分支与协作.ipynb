# Git 分支与协作

## 为什么需要它

Git **创建分支只是写了一个 41 字节的指针文件，不拷贝任何代码。** 

## 是什么

### 分支的本质

Git 的分支是一个**指向某个 commit 的可移动指针**。main 是一个指针，feature 是另一个指针，它们指向各自的 commit。

`HEAD` 是一个特殊指针，指向「你当前在哪个分支上」。切换分支就是移动 HEAD。

````python
        main (指针)
         │
    A───B───C
             │
         feature (指针)
````

### 核心概念

| 概念 | 一句话 |
|------|--------|
| **branch** | 指向某个 commit 的指针，一条独立的工作线 |
| **merge** | 把 A 分支的 commit 合入 B 分支，产生一个合并提交 |
| **rebase** | 把当前分支的 commit「摘下来」逐个放到目标分支顶端，历史变直线 |
| **remote** | 远程仓库的简称。克隆（clone）时 Git 自动把来源地址命名为 `origin` |
| **push** | 把本地 commit 上传到远程仓库 |
| **pull** | 把远程的 commit 下载并合并到本地（= fetch + merge） |
| **fetch** | 只下载远程 commit，不动本地工作目录 |
| **stash** | 临时把工作区的修改压入一个栈，之后再取出来 |

> **origin**：执行 `git clone <url>` 时，Git 自动把那个 URL 记为一个叫 `origin` 的远程仓库简称。之后 `git push origin main` 就是「把本地 main 推送到 origin 那个地址」。

## 怎么样

### 分支操作

````python
# 只创建分支
git branch feature-login

# 创建并切换到该分支
git checkout -b feature-login

# 查看所有分支（当前分支前有 *）
git branch

# 删除分支
git branch -d feature-login
````

### merge —— 合并

把 feature 分支上所有的 commit 合入 main。Git 会自动找出两个分支的共同祖先，把 feature 独有的 commit 应用到 main 上，最后生成一个**合并提交（merge commit）**——这个 commit 有两个父 commit，记录「我是从哪两条线合并来的」。

```text
合并前：                           合并后：
      main    A───B───C               main    A───B───C───────M
                      │                             \      /
      feature         D───E           feature         D───E
```

当两个分支改了同一文件的同一行时，Git 无法判断留哪个——这叫**冲突（conflict）**。Git 会在文件里同时标出两个版本，由人来决定。

````python
# 切换到 main 并合并 feature
git checkout main
git merge feature-login

# 如果冲突了，手动编辑冲突文件，然后：
git add .
git commit -m "解决合并冲突"
````

### rebase —— 变基

和 merge 目标相同（合入改动），但手段不同：rebase 把当前分支独有的 commit 一个个摘下来，接到目标分支顶端。结果是一条直线历史，没有分叉的合并提交。

```text
变基前：                           变基后：
      main    A───B───C               main    A───B───C
                      │                             │
      feature         D───E           feature         D'──E'
```

> ⚠ 不要对已经 push 到远程的分支做 rebase。rebase 会改写 commit 哈希（D→D'），别人本地还留着旧的 D，两边历史不一致，下次 push 会乱。

````python
git checkout feature-login
git rebase main

# 有冲突时解决后继续
git rebase --continue
# 或放弃
git rebase --abort
````

### 远程协作

**远程仓库**（remote）是存在另一台机器上的 Git 仓库副本。`git clone` 把远程仓库完整复制到本地，并自动将来源地址命名为 `origin`。

````python
# 克隆远程仓库
git clone git@github.com:user/repo.git

# 查看关联的远程仓库
git remote -v

# 推送本地 commit 到远程
git push origin main

# 首次推送并绑定上游（之后 git push 即可）
git push -u origin feature-login

# 拉取远程 commit 并合并到本地（= fetch + merge）
git pull

# 只下载不合并（先看看远程有什么变化）
git fetch origin
````

### stash —— 暂存

改到一半要切分支修 bug 但不想 commit 半成品？`stash` 把工作区和暂存区的改动压进一个临时栈，切回来时 `pop` 恢复。

````python
git stash        # 储藏当前修改
git stash list   # 查看储藏列表
git stash pop    # 取出最新储藏并删除
git stash apply  # 取出但不删除
````
