# Shell 与终端基础

> bash / zsh / PATH / 环境变量

## 为什么需要它

1980 年代，操作计算机的唯一方式是命令行——打一条命令，计算机执行，返回文字结果。鼠标和图形界面（GUI）是后来才出现的。

到今天，图形界面能做大部分日常操作（打开文件、看网页），但有些事情鼠标永远做不了：**批量重命名一千个文件、每天凌晨自动备份数据库、同时在一百台服务器上部署代码**——这些需要**脚本化**和**自动化**，而命令行（Shell）就是做这个的界面。

对开发者来说，Git、Docker、Python、服务器运维全都运行在 Shell 里。你可以不用 GUI，但不可能不用 Shell。

## 是什么

三个容易混的概念，一次剥开：

```text
┌─────────────────────────────────────────┐
│  终端模拟器（Terminal Emulator）          │ ← 你看到的窗口软件
│  ┌───────────────────────────────────┐  │    (Windows Terminal, iTerm2)
│  │  Shell（bash / zsh / fish）        │  │ ← 解析命令、运行程序
│  │  ┌─────────────────────────────┐  │  │
│  │  │ 你输入 ls -la               │  │  │
│  │  │ Shell 找到 /bin/ls 并运行    │  │  │ ← 这一层是人机翻译
│  │  │ 结果打印回终端                │  │  │
│  │  └─────────────────────────────┘  │  │
│  └───────────────────────────────────┘  │
└─────────────────────────────────────────┘
```

| 概念 | 说明 |
|------|------|
| **终端模拟器** | 窗口软件。负责显示、字体、颜色 |
| **Shell** | 命令解释器。bash 和 zsh 是两种最流行的 Shell |
| **bash** | 大多数 Linux 默认 Shell，诞生于 1989 年 |
| **zsh** | macOS 默认 Shell，兼容 bash，额外提供更好的自动补全和主题 |

### PATH —— 去哪找命令

`PATH` 是 Shell 里最重要的环境变量。它存了一串目录路径，当你打 `git` 时，Shell 按顺序在这些目录里找名叫 `git` 的可执行文件，找到第一个就运行。

```shell
$ echo $PATH
/usr/local/bin:/usr/bin:/bin:/home/user/.local/bin

打 git → 找 /usr/local/bin/git（没有）→ 找 /usr/bin/git（找到了！运行）
```

`command not found` = 在 PATH 所有目录里都没找到这个可执行文件。要么没装，要么不在 PATH 里。

### 环境变量 —— 给程序传配置

**环境变量**是操作系统级的 key-value，每个程序启动时自动继承。`DATABASE_URL`、`SECRET_KEY` 这些敏感信息通常不写死在代码里，而通过环境变量传入——不同环境（开发/测试/生产）设不同的值，代码不用改。

## 怎么样

### 常用命令

````python
pwd                  # 当前目录（print working directory）
ls                   # 列出文件
ls -la               # 详细列表（权限、大小、隐藏文件）
cd /path/to/dir      # 切换目录
cd ..                # 上一级
cd ~                 # 回家目录

cat file.txt         # 打印文件内容
head -20 file.txt    # 看前 20 行
tail -f app.log      # 实时跟踪日志末尾

cp src dst           # 复制
mv old new           # 移动/重命名
rm file.txt          # 删除（无回收站！）
rm -rf dir/          # 递归强制删除（极度危险）
mkdir newdir         # 新建目录

echo "hello"         # 输出文本
which python         # 查看命令完整路径
````

### 管道与重定向

Shell 真正强大的核心：**把小命令像积木一样拼接**。

- **管道** `|`：左边命令的输出变成右边命令的输入
- **重定向** `>`：把输出写到文件（覆盖），`>>` 追加到末尾，`<` 从文件读取

````python
# 管道：输出 → 过滤 → 计数
cat app.log | grep ERROR | wc -l

# 重定向
echo "hello" > file.txt     # 覆盖写入
echo "world" >> file.txt    # 追加
python script.py < input.txt # 从文件读取输入
````

### 环境变量操作

````python
echo $PATH              # 查看
echo $HOME

export MY_VAR="hello"    # 临时设置（仅当前会话）

DEBUG=true python app.py # 仅对这一次命令有效
````

### 配置文件

每次打开终端，Shell 自动加载配置文件。改了 `PATH` 一关终端就没了？写到配置文件里。

| Shell | 登录时执行 | 每次开终端执行 |
|-------|-----------|-------------|
| bash | `~/.bashrc` | `~/.bashrc` |
| zsh | `~/.zshrc` | `~/.zshrc` |

### Shell 脚本

把多个命令写进 `.sh` 文件，顶部 `#!/bin/bash`（叫 **shebang**，告诉系统用哪个解释器运行），赋予执行权限后就可以像程序一样调用。

````python
#!/bin/bash
# shebang：指定用 bash 执行

DATE=$(date +%Y%m%d)
tar -czf "backup-$DATE.tar.gz" /path/to/data
echo "备份完成: backup-$DATE.tar.gz"
````
