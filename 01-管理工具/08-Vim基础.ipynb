## 为什么需要它

通过 **SSH**（Secure Shell，一种加密的远程连接协议）登录到远程服务器后，面对的是一个纯黑终端窗口——没有桌面，没有鼠标，没有 VS Code。改一个配置文件、写一段脚本，只能靠终端里自带的编辑器。

`nano` 最简单（底部有快捷键提示，上手零成本）但功能少。`vim` 几乎每台 Unix/Linux/macOS 都预装，功能强大，编辑效率极高——但前提是你得先学会它的操作逻辑。

对开发者来说，不需要变 vim 高手，但至少要掌握生存技能：打开文件、编辑、保存、退出。不然连服务器上的 nginx 配置都改不了。

## 是什么

**Vim**（Vi IMproved）是 1991 年在 vi 基础上改进的终端文本编辑器。其核心设计和其他所有编辑器都不一样：**模式（Mode）**。

```text
            按 i / a / o             按 Esc
           ─────────────>          <───────────
  普通模式                     插入模式
  (Normal)                    (Insert)
  移动光标、                   正常打字
  复制粘贴、
  删除、撤销                      按 :
                              ──────>
                              命令行模式
                              (Command)
                              保存、退出、搜索、替换
```

新手打开 vim 想打字，发现键盘没反应——因为你处在「普通模式」，要先按 `i` 才能进入「插入模式」。这就是 vim 著名的「新手出不去」体验。

模式设计的目的：手不离开主键盘区就能完成所有操作。一旦熟练，编辑速度远超鼠标操作。

### neovim

**neovim** 是 vim 的现代化分支，保留了全部核心编辑功能，重构后支持更好的插件生态和内置 **LSP**（Language Server Protocol，编辑器提供代码补全/跳转/错误提示的后台协议）客户端。如今大多数 vim 用户实际在用 neovim。

## 怎么样

### 生存必备

````python
vim file.txt          # 打开/创建文件

i        # 光标前插入（insert）
a        # 光标后插入（append）
o        # 下一行插入（open line）
Esc      # 退出插入模式，回到普通模式

:w       # 保存（write）
:q       # 退出（quit）
:wq      # 保存并退出
:q!      # 强制退出，不保存
````

### 光标移动（普通模式）

Vim 用 `hjkl` 代替方向键，手不离开主键盘：

```text
        k (上)
h (左)         l (右)
        j (下)
```

| 快捷键 | 跳转到 |
|--------|--------|
| `w` | 下一词开头（word） |
| `b` | 上一词开头（back） |
| `0` | 行首 |
| `$` | 行尾 |
| `gg` | 文件开头 |
| `G` | 文件末尾 |

### 编辑命令（普通模式）

````python
x        # 删除光标处字符
dd       # 删除（剪切）整行
yy       # 复制整行（yank）
p        # 粘贴
u        # 撤销
Ctrl+r   # 重做

# 组合：动词 + 范围 = Vim 编辑语法的核心
dw       # 删除一个单词（delete word）
d$       # 删到行尾
ciw      # 修改光标处单词（change inner word）
di(      # 删除括号内全部内容
````

### 搜索与替换

````python
# === 搜索 ===
/keyword     # 向下搜索
?keyword     # 向上搜索
n            # 下一个匹配
N            # 上一个匹配

# === 替换（命令行模式）===
:%s/old/new/g     # 全文替换
:%s/old/new/gc    # 全文替换，逐个确认
:5,10s/old/new/g  # 仅 5-10 行替换
````

### 视觉模式

普通模式按 `v` → 移动光标选中文本（纯键盘拖选）→ `d` 删除 / `y` 复制 / `>` 缩进。

### 速记卡

```text
vim file   打开        i        编辑
Esc        退出编辑    :w       保存
:q         退出        :wq      保存退出
:q!        强制退出    /word    搜索
u          撤销        dd       删一行
yy         复制一行    p        粘贴
```

12 条命令，覆盖 90% 服务器编辑需求。
