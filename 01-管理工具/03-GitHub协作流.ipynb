# GitHub 协作流

> PR / issue / actions / fork / .gitignore

## 为什么需要它

**GitHub**：在 Git 仓库之上盖了一层「协作操作系统」。代码托管、Pull Request（提修改→审查→合并的标准化流水线）、Issue（任务追踪）、Actions（自动化）、Fork（跨账号协作）——全都围绕一个 Git 仓库可视化地运转起来。

## 是什么

GitHub 是一个代码托管平台，核心资产是围绕 Git 仓库的一整套协作工具：

| 功能 | 解决的问题 |
|------|-----------|
| **仓库托管** | 代码存在云端，有一个团队公认的权威副本 |
| **Pull Request（PR）** | 把「请合并我的代码」变成标准化流程：提交→审查→讨论→合并 |
| **Issue** | 报 bug、提需求、讨论方案——每个任务一个编号，可引用、可自动关闭 |
| **Actions（CI/CD）** | 事件触发自动化：push 自动跑测试、PR 自动检查、合并自动部署 |
| **Fork** | 把别人的仓库完整复制到自己账号下，改完提 PR 回原仓库 |
| **.gitignore** | 告诉 Git 哪些文件不要管（编译产物、依赖包、密钥……） |

> **CI/CD**（持续集成/持续部署）：代码一提交就自动构建、测试（CI），测试通过自动部署到服务器（CD）。GitHub Actions 是 GitHub 内置的 CI/CD 引擎。

## 怎么样

### Pull Request —— 核心协作机制

PR 把一次代码合并变成五步流程：

````python
1. 从 main 拉出 feature 分支，写代码、commit、push
2. 在 GitHub 网页创建 PR：选 base（往哪合）和 compare（你的分支），写描述
3. 队友在 PR 页面上逐行审查（Code Review），评论、点 Approve 或 Request Changes
4. 如果有 CI 配置，GitHub Actions 自动跑测试、类型检查
5. 全部通过后点 Merge → 代码合入 main
````

> PR 是对 `git merge` 的包装。点 Merge 按钮背后执行的还是 `git merge`。PR 的价值在合并之前的**审查、讨论、自动化检查**。

### Issue —— 任务原子单位

每个 Issue 有唯一编号（#1, #2...），可以在 commit 消息或 PR 描述里引用。PR 描述里写 `fixes #42`，合并后 #42 自动关闭。Issue 还可以打 Label（分类标签）、设 Milestone（里程碑）、指派负责人。

### Fork —— 给无权限者开一扇门

你没权写别人的仓库，但可以 **Fork**——把自己的账号当作中转站：原仓库 → Fork 到你名下（你有全部权限）→ 在你的副本里改 → 从你的副本向原仓库提 PR。整个开源世界（包括 Linux 内核的数万贡献者）都靠这套机制运转。

### Actions —— 自动化流水线

在 `.github/workflows/` 目录放 **YAML**（一种人类可读的配置格式，见 [03-编码与序列化/YAML与TOML.ipynb](../03-编码与序列化/YAML与TOML.ipynb)）文件定义规则：push 时跑什么、PR 时检查什么。GitHub 提供免费的运行环境来执行。

### .gitignore

把不应进 Git 的文件名或模式写入 `.gitignore`，Git 自动跳过它们。

````python
__pycache__/      # Python 编译缓存
*.pyc             # 编译后的 pyc 文件
.venv/            # 虚拟环境
node_modules/     # JS 依赖包
.env              # 环境变量（含密钥，绝对不能提交）
dist/             # 构建产物
.DS_Store         # macOS 系统文件
*.log             # 日志
````

## 完整协作流程

```text
创建 Issue #42「登录页需加微信登录」
  → git checkout -b feat/wechat-login
  → 写代码、commit、push
  → 在 GitHub 创建 PR，描述里写 fixes #42
  → Actions 自动跑测试和 lint
  → 同事逐行 review，在第 35 行提修改建议
  → 你改好再 push，Actions 重跑
  → 同事点 Approve，你点 Merge
  → #42 自动关闭，Actions 检测到 main 有更新 → 自动部署
```

这就是一个 feature 从想法到上线的完整链路。
