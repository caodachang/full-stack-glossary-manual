# Git 概述与基础操作

## 为什么需要它

**SVN**（一种集中式版本控制工具）。把所有代码和版本历史存在一台中央服务器上，需要先联网才能使用。

**Git**：把所有代码和版本历史存在自己电脑上。想跟别人同步只需要 push/pull。

## 是什么

**Git** 是一个分布式版本控制系统。核心工作是：**跟踪文件每次修改，存成快照，让你随时回到任意历史版本。**

### 三个区域

Git 里文件经过三个区域才完成一次「保存」：


```python
工作目录              暂存区               本地仓库
(Working             (Staging            (Repository)
 Directory)           Area / Index)       (.git 目录)
    │                   │                    │
    │   git add         │   git commit       │
    │ ─────────────────>│ ──────────────────>│
    │                   │                    │
    │<──────────────────│                    │
    │   git restore     │                    │
```

- **暂存区**：一个中间区域，用来挑选哪些修改放入下一次 commit。
- **本地仓库**：就是项目根目录下的 **`.git`** 隐藏文件夹。

### 每次 commit 存什么

每次 `git commit`，Git 对当前暂存区里的所有文件拍一张**快照**（snapshot），存进 `.git`。每个 commit 附带：
- 作者是谁、什么时候提交的
- 一段消息说明做了什么变动
- 一个**哈希值**作为唯一标识，没改过的文件哈希值不会变只会引用以前的。

## 怎么样

下面 5 个命令构成日常最高频的循环。

### git init —— 初始化

把一个普通目录变成 Git 仓库，就是在目录里创建 `.git` 文件夹。

```python
# 初始化一个新仓库
git init

# 查看隐藏的.git目录
ls -la .git/       
```

### git status —— 查看状态

回答一个问题：「现在仓库什么状态？」——哪些文件改了没 add？哪些 add 了没 commit？

```python
git status  # 红色是未追踪的文件，绿色是暂存区内的文件
```

### git add —— 添加到暂存区

把工作目录中的修改加入暂存区。

```python
# 添加单个文件
git add README.md

# 添加当前目录下所有修改
git add .
```

### git commit —— 提交

把暂存区的内容拍一张快照，永久存入 `.git`。`-m` 后面写消息说明这个 commit 做了什么。

```python
# 提交并写消息
git commit -m "添加 README 文件"

# -a 跳过 git add，把所有已跟踪文件（ add 或 commit过的文件）的修改直接提交
git commit -am "快速提交"
```

### git log —— 查看历史

按时间倒序显示所有 commit。`--oneline` 让每个 commit 只占一行，`--graph` 用 ASCII 画出分支线。

```python
# 查看完整历史
git log

# 紧凑格式
git log --oneline

# 图形化分支历史 --all 显示所有分支
git log --oneline --graph --all
```

### git diff —— 查看差异

告诉你两份内容到底哪几行不一样。三种对比场景：
- `git diff`：工作目录 vs 暂存区（改了什么还没 add）
- `git diff --staged`：暂存区 vs 最新 commit（add 了什么还没 commit）
- `git diff HEAD`：工作目录 vs 最新 commit（所有还没 commit 的变动）

```python
# 查看未暂存的修改
git diff

# 查看已暂存的修改
git diff --staged

# 查看与上次提交的区别
git diff HEAD
```
