# PATH与环境变量

## PATH是什么

给 Shell 找命令用的目录列表

## 环境变量是什么

给程序看的会变化的运行配置，例如：

| 变量名 | 可能含义 |
| --- | --- |
| APP_ENV | 当前运行环境，比如 development 或 production |
| PORT | 程序监听端口 |
| DATABASE_URL | 数据库连接地址 |
| API_KEY | 第三方服务密钥 |
| PATH | 命令查找目录列表 |

## 用户级、系统级、项目级

环境变量也有作用范围。

| 层级 | 影响范围 | 例子 |
| --- | --- | --- |
| 系统级 | 整台机器、所有用户 | 系统 PATH |
| 用户级 | 当前用户 | 用户自己的工具路径 |
| 终端会话级 | 当前打开的终端窗口 | 临时设置一个变量 |
| 项目级 | 当前项目 | .env 文件、本地开发配置 |
| 部署环境级 | 某个线上服务 | 云平台环境变量 |

## 在规格里怎么写

项目涉及部署、第三方服务、数据库、命令行工具，就应该写清环境变量要求。

更好的写法是：

| 变量名 | 用途 | 是否必填 | 示例 / 说明 |
| --- | --- | --- | --- |
| APP_ENV | 运行环境 | 是 | development、production |
| DATABASE_URL | 数据库连接地址 | 是 | 由部署平台注入 |
| API_KEY | 第三方服务密钥 | 是 | 不提交到仓库 |
| PORT | 服务端口 | 否 | 默认 3000 |
