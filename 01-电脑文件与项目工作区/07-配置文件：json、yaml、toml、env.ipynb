# 配置文件：json、yaml、toml、env

## JSON

**JSON** 是一种结构化数据格式，常见于前端、API、配置文件。

示例：

```json
{
  "name": "demo-app",
  "version": "1.0.0",
  "scripts": {
    "dev": "vite",
    "test": "vitest"
  }
}
```

JSON 常见场景：

- package.json
- API 返回数据
- 前端配置

## YAML

**YAML** 更强调人类可读性，常用于部署、CI/CD、配置流程。

特点：

- 支持注释
- 对缩进敏感

示例：

```yaml
name: ci
on:
  - push
jobs:
  test:
    runs-on: ubuntu-latest
```

## TOML

**TOML** 是一种适合人手写的配置格式。

特点：

- 支持注释
- 结构比 JSON 清爽
- Python 项目里很常见

示例：

```toml
[project]
name = "demo"
version = "0.1.0"
dependencies = ["requests"]

[tool.pytest]
testpaths = ["tests"]
```

## env

**env 文件**通常用来保存环境变量，不适合保存复杂嵌套结构。

它不像 JSON、YAML、TOML 那样强调复杂结构，而是简单的 key-value：

```dotenv
DATABASE_URL=postgres://example
API_KEY=your-api-key
APP_ENV=development
```

## 在规格里怎么写

如果项目需要配置，就应该在规格里写清楚配置要求。

可以写成表格：

| 配置项 | 格式 / 位置 | 用途 | 是否提交 |
| --- | --- | --- | --- |
| DATABASE_URL | .env | 数据库连接地址 | 不提交真实值 |
| API_KEY | .env | 第三方服务密钥 | 不提交真实值 |
| package.json | JSON | 前端依赖和脚本 | 提交 |
| pyproject.toml | TOML | Python 项目配置 | 提交 |
| CI 配置 | YAML | 自动测试和构建 | 提交 |

也可以写规则：

- 真实密钥只放环境变量，不写进代码
- 仓库提交 .env.example，不提交 .env
- README 需要说明如何准备配置
- 新增配置项时，同时更新示例文件和文档
