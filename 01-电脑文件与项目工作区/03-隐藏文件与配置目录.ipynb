# 隐藏文件与配置目录

## 隐藏文件长什么样

在 macOS / Linux 和很多开发工具里，名字以点开头的文件或目录，经常会被当作隐藏文件。

在 Windows 里，文件也可以通过“隐藏属性”隐藏，不一定非要点开头。

## 配置目录是什么

**配置目录**就是专门存配置的一类目录。

| 层级 | 影响范围 | 例子 |
| --- | --- | --- |
| 用户级配置 | 当前用户的多个项目 | .ssh、.gitconfig |
| 项目级配置 | 当前项目 | .vscode、.env、.gitignore |
| 工具缓存 | 工具运行时生成 | .cache、构建缓存 |

## 在规格里怎么写

项目涉及**配置、密钥、环境变量、部署**，就应该写清楚配置管理方式。

例如：

配置管理：

- 本地开发使用 .env 保存环境变量
- 仓库提交 .env.example，不提交真实 .env
- 生产环境通过部署平台注入环境变量
- .gitignore 需要忽略真实密钥、缓存、日志和本地虚拟环境
