# 项目文件夹里通常有什么

## 常见项目结构

不同语言和框架会有不同结构，但常见类别差不多。

| 类别 | 常见名称 | 作用 |
| --- | --- | --- |
| 说明文档 | README.md、docs | 说明项目是什么、怎么运行 |
| 源码目录 | src、app、pages | 放主要代码 |
| 测试目录 | tests | 放测试文件 |
| 静态资源 | assets、public、static | 图片、字体、图标等 |
| 配置文件 | package.json、pyproject.toml、tsconfig.json | 告诉工具如何处理项目 |
| 依赖锁文件 | package-lock.json、pnpm-lock.yaml、uv.lock | 固定依赖版本 |
| 环境变量示例 | .env.example | 说明需要哪些环境变量 |
| 忽略规则 | .gitignore | 告诉 Git 哪些文件不提交 |
| 构建产物 | dist、build | 工具生成的发布文件 |

## README通常是入口

打开陌生项目，第一眼应该找 README。

README 通常回答：

- 这个项目是什么？
- 适合谁用？
- 怎么安装依赖？
- 怎么启动？
- 怎么测试？
- 有哪些重要约定？

## 源码目录放主要逻辑

常见名字：

| 名称 | 常见场景 |
| --- | --- |
| src | 通用源码目录 |
| app | 应用主目录，很多 Web 框架常见 |
| pages | 页面路由相关目录 |
| components | 前端组件目录 |
| lib | 通用库代码 |
| services | 服务或接口相关代码 |

## 配置文件告诉工具怎么工作

常见例子：

| 文件 | 常见用途 |
| --- | --- |
| package.json | 前端 / Node 项目配置、脚本、依赖 |
| pyproject.toml | Python 项目配置、依赖、工具设置 |
| tsconfig.json | TypeScript 配置 |
| vite.config.js | Vite 构建配置 |
| eslint.config.js | 代码检查规则 |
| .env.example | 环境变量示例 |

配置文件的价值是：**让工具知道这个项目应该怎么运行、检查、构建和发布**。

## 在规格里怎么写

但如果你让 AI 创建一个新项目，或者让别人搭建项目骨架，可以写清楚基本结构。

这样 AI 生成项目时，不容易把文件乱放。
