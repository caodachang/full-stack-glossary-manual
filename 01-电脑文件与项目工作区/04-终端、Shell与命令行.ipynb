# 终端、Shell与命令行

| 概念 | 它是什么 | 例子 |
| --- | --- | --- |
| Shell | 解释命令的程序 | PowerShell、CMD、bash |
| CLI | Command Line Interface，命令行接口 | git、npm、uv、docker |

## 在规格里怎么写

项目涉及**开发环境、CLI 工具、测试、构建、部署、AI Agent 执行步骤**，就应该把命令相关信息写清楚。

例如：

- 在项目根目录运行开发命令
- 成功后终端输出本地访问地址
- 测试命令失败时显示失败用例和原因
- Windows 使用 PowerShell 示例，macOS / Linux 使用 bash 示例
