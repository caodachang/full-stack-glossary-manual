# 文件、目录与路径

## 文件是什么

不同文件类型的区别在于**扩展名**和它里面保存的**数据格式**。

常见例子：

~~~text
README.md        一份 Markdown 文档
main.py          一份 Python 代码文件
index.html       一个 HTML 页面
logo.png         一张图片
package.json     一个前端项目配置的JSON文件
.env             一份环境变量配置文件
app.log          一份日志文件
~~~

开发时要特别注意：文件名不是随便起的。很多工具会根据固定文件名寻找配置，例如 package.json、pyproject.toml、.gitignore、README.md。

## 目录

.:当前目录

..:上级目录

## 在规格里怎么写

涉及文件上传、导入导出、静态资源、配置文件、生成文件，需要把文件位置或文件规则写清楚。

例如文件上传功能可以写：

~~~text
功能：用户上传头像
文件类型：jpg、png、webp
文件大小：最大 5MB
存储位置：对象存储的 user-avatar 目录
访问方式：上传成功后返回头像 URL
错误处理：文件类型不支持时提示“仅支持 jpg、png、webp”
~~~

注意：规格里常写上传目录、对象存储路径、导出文件命名规则、配置文件位置。

## 面试常见问法

### 为什么程序会报“找不到文件”？

可以从几个方向回答：

- 当前工作目录不对
- 相对路径写错
- 文件实际不存在
- 文件名大小写不一致
- 部署环境和本地环境目录结构不同
- 权限不足导致无法读取
