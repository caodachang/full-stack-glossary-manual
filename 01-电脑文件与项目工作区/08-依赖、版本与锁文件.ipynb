# 依赖、版本与锁文件

## 依赖分几类

不是所有依赖都在同一个阶段使用。

| 类型 | 什么时候需要 | 例子 |
| --- | --- | --- |
| 运行依赖 | 程序真正运行时需要 | react、fastapi、requests |
| 开发依赖 | 开发、检查、测试时需要 | eslint、pytest、black、vitest |
| 构建依赖 | 打包或编译时需要 | vite、webpack、typescript |
| 可选依赖 | 有某功能才需要 | 图片处理、数据库驱动、特定平台包 |

- 线上环境通常只需要运行依赖
- CI 环境通常需要测试和构建依赖
- 本地开发环境可能需要更多工具依赖
- 依赖越多，安装越慢，安全风险也越大

## 最小示例

前端项目的依赖清单可能长这样：

```json
{
  "dependencies": {
    "react": "^18.2.0"
  },
  "devDependencies": {
    "vite": "^5.0.0",
    "vitest": "^1.0.0"
  }
}
```

Python 项目的依赖清单可能长这样：

```toml
[project]
dependencies = [
  "requests>=2.31.0",
  "fastapi>=0.110.0"
]

[dependency-groups]
dev = [
  "pytest>=8.0.0"
]
```

## 在规格里怎么写

但如果项目涉及工程交付，至少应该写清这些约束：

| 项目 | 规格里应该写到什么程度 |
| --- | --- |
| 语言版本 | 例如 Python 3.12、Node 20 |
| 包管理器 | 例如 uv、npm、pnpm |
| 关键框架 | 例如 React、FastAPI、Next.js |
| 锁文件策略 | 是否提交锁文件，是否要求可复现安装 |
| 安全要求 | 是否需要依赖扫描，是否禁止已知高危依赖 |

## 面试常见问法

### package.json 和 package-lock.json 有什么区别？

package.json 声明项目需要哪些依赖和脚本；package-lock.json 记录实际安装出的精确依赖树，用来提高环境复现能力。

### ^ 和 ~ 有什么区别？

大致可以理解为：`^` 通常允许较宽的兼容升级，`~` 通常更保守，主要允许补丁级升级。具体规则要看语言生态和包管理器。

## 常见坑

- **看到版本范围就以为一定安全**：范围只是规则，仍然要测试。
