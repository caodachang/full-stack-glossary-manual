# 文件类型与扩展名

## 常见文件类型速览

| 类型 | 常见扩展名 | 主要用途 |
| --- | --- | --- |
| 文本文档 | .txt、.md | 普通文本、说明文档、笔记 |
| 代码文件 | .py、.js、.ts、.java、.go | 保存源代码 |
| 网页文件 | .html、.css、.js | 页面结构、样式、交互 |
| 数据文件 | .json、.yaml、.toml、.csv | 保存配置或结构化数据 |
| 图片文件 | .png、.jpg、.webp、.svg | 图片、图标、矢量图 |
| 压缩文件 | .zip、.tar、.gz、.7z | 打包和压缩 |
| 可执行文件 | .exe、.msi、.bat、.sh | 程序、安装包、脚本 |
| Notebook | .ipynb | Jupyter / VS Code notebook 文档 |
| 日志文件 | .log | 程序运行记录 |

## 文件类型和扩展名不是一回事

文件类型可以由多种方式判断：

- **扩展名**：看文件名后缀，例如 .png
- **文件内容格式**：看文件内部数据结构（如：JSON，CSV）
- **MIME Type**：网络传输时说明内容类型
- **操作系统关联**：默认打开方式

例如一个图片文件可能同时有这些信息：

- 文件名：logo.png
- 扩展名：png
- MIME Type（网络世界里的类型说明）：image/png
- 默认打开程序：照片查看器或浏览器

## 在规格里怎么写

只要功能涉及上传、下载、导入、导出、附件、图片、报表，就应该写清楚文件类型。

例如头像上传：

功能：上传用户头像
允许类型：jpg、png、webp
最大大小：5MB
最小尺寸：200x200
保存方式：上传到对象存储
返回结果：头像 URL
错误提示：文件类型不支持、文件过大、上传失败

## 面试常见问法

### 扩展名和文件格式有什么区别？

回答：

扩展名是文件名里的提示，文件格式是文件内容的组织方式。
扩展名可以改，但内容格式不会因为改名自动改变。
