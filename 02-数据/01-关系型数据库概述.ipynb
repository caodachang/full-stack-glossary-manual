# 关系型数据库概述

## 关系型数据库一切数据都在表里

**SQLite** 整个数据库就是一个文件。

**MySQL** 特点是读快、简单、生态广。

**PostgreSQL** 是功能最强的开源关系型数据库。

**Oracle** 是商业数据库之王。银行、政府、电信公司的核心系统跑 Oracle。

**SQL Server** 是微软的关系型数据库。

> **SQLite** 单体项目 → **MySQL** 读多写少的 Web 应用 → **PostgreSQL** 功能优先的复杂应用 → **Oracle/SQL Server** 不差钱的企业。

## 核心概念：表、SQL、ACID、范式

### 表（Table）与关系

表像 Excel 工作表——每行是一条记录，每列是一个字段。

### SQL：操作数据库的统一语言

分成四个子语言
- DDL：定义结构
- DML：操作数据
- DQL：查询（DML 的子集）
- DCL：控制权限

### ACID：数据库的"靠谱"保证

* **A（原子性）**：多步骤要么全做，要么全不做。
* **C（一致性）**：事务前后，所有数据约束仍然成立。
* **I（隔离性）**：并发事务互不干扰。
* **D（持久性）**：把数据写到磁盘（非易失存储）才算完成。

### 范式：如何设计表结构

**第一范式（1NF）**：每列存的是原子值。

**第二范式（2NF）**：非主键列必须完全依赖主键（不重复的列）。

**第三范式（3NF）**：非主键列不能依赖其他非主键列。

## 本篇出现的新名词速查

| 名词 | 一句话解释 |
|------|-----------|
| 关系型数据库 (RDBMS) | 用表（关系）组织数据、用 SQL 操作、满足 ACID 的数据库系统 |
| SQL | Structured Query Language——操作关系型数据库的声明式语言 |
| 表 (Table) | 数据库存数据的基本单位，行=记录，列=字段 |
| ACID | 数据库四保证：原子性、一致性、隔离性、持久性 |
| 范式 (Normal Form) | 设计表结构的规则，目的是减少冗余和更新异常 |
| JOIN | 把多张表的关联行拼成一张结果集 |
| 主键 (Primary Key) | 唯一标识表中每一行的列，不能重复不能为空 |
| 外键 (Foreign Key) | 一个表的列指向另一个表的主键，建立关联 |
