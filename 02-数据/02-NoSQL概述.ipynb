# NoSQL 概述

## 历史动机：为什么会有 NoSQL？

Web 2.0 从"只读"到"读写"的转变，传统关系型数据库暴露了弱点

出现了 **NoSQL**（Not Only SQL），是一类**不用关系模型**的数据库的统称。

## NoSQL 是什么？

NoSQL 是四种不同的数据模型的统称。

### 文档型

**文档型数据库**存的是**自包含的 JSON 式文档**。一个用户的所有信息全放在一个文档里，不需要 JOIN，一次查询拿出全部。**MongoDB** 是这个类型的代表。

### 键值型

**键值型数据库**就是一个巨大的字典，`key → value`。你给一个键，它立刻返回对应的值。

### 列族型

**列族型数据库**把数据按列存储而不是按行。查到一行后，只读出需要的列，不碰其他列。**Cassandra** 是这个类型的代表，无单点故障，跨数据中心天然支持。

### 图型

**图型数据库**存的是实体和实体之间的关系。"你朋友的朋友也买了什么"在图数据库里几次指针跳转就出来。**Neo4j** 是代表。

> 四种类型的共同点：都**放弃了关系模型**，换来的是水平扩展、灵活 Schema、和特定场景下的极致性能。

## 本篇出现的新名词速查

| 名词 | 一句话解释 |
|------|-----------|
| NoSQL | Not Only SQL——不用关系模型的数据库统称，四大类型 |
| MongoDB | 文档型 NoSQL 代表，存 JSON 式文档，最接近 SQL 思维 |
| Redis | 键值型 NoSQL 代表，内存中运行，五种数据结构，缓存之王 |
| Cassandra | 列族型 NoSQL 代表，无单点故障，跨数据中心，海量写入 |
| Neo4j | 图型 NoSQL 代表，多跳关系查询秒出，反欺诈和推荐系统用 |
| 最终一致 | 数据最终会一致，中间允许短暂不一致窗口 |
| 文档型 | 存自包含的 JSON 式文档，不需要 JOIN |
| 键值型 | 存巨大的 key→value 字典，极简极快 |
| 列族型 | 按列存数据，海量写入场景 |
| 图型 | 存实体和关系，多跳查询秒出 |
