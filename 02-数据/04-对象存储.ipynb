# 对象存储

## 对象存储是什么？

把存储抽象成一个无限大的空间——你只管往里丢文件，不用管它在哪台机器上、怎么备份、怎么扩容量。文件不再叫"文件"，叫**对象**。对象存储的三个核心概念：**Bucket**（桶——存对象的容器）、**Object**（对象——文件本体 + 元数据 + 唯一 ID）、**Key**（键——对象的路径名）。

对象存储和传统文件系统最大的区别：**没有目录树**。

## 典型使用场景

### 静态资源托管

网站的所有图片、CSS、JS、字体文件放在对象存储里，直接通过 HTTPS 访问。前端请求 HTML 从服务器走，但静态资源全从对象存储拉。

### 数据湖

把公司所有数据全部原样丢进对象存储，不预先清洗，不预先建表。需要分析时，大数据引擎直接查询对象存储上的文件。这种"先存后治"的模式叫**数据湖**。

## 本篇出现的新名词速查

| 名词 | 一句话解释 |
|------|-----------|
| 对象存储 | 无限扩展的扁平 key→object 存储，HTTP 直接访问，无目录树 |
| S3 | 对象存储的事实标准 |
| Bucket | 桶——S3 里存对象的容器，每个桶有全局唯一的名字 |
| Object | 对象——文件本体 + 元数据（类型、权限） + 唯一 ID |
| Key | 对象的路径名，`/photos/beach.jpg` 就是一整个 key，不是目录 |
| 块存储 | 裸硬盘空间，建文件系统，数据库低延迟需求用 |
| 文件存储 | 有目录树的共享文件系统，NFS 挂载，文件多了慢 |
| 数据湖 | 各类数据原样丢进对象存储，分析时再查询——先存后治 |
