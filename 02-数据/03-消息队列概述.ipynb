# 消息队列概述

## 消息队列是什么？

消息队列解决：**异步**（把不紧急的事往后放）和**削峰**（请求先堆在队列里，后端按自己节奏处理）。

## 核心概念

### 三种消息模式

*   **点对点**：一条消息只被一个消费者处理。
*   **发布订阅**：一条消息所有订阅者都收到。
*   **请求回复**：生产者发一条消息后等消费者的回复。

### 消息可靠性

*   **生产者确认**：队列收到消息后回复"我收了"（ack）。
*   **消费者确认**：消费者处理完后回复 ack。
*   **死信队列**：一条消息被反复重试仍然失败，进入死信队列等待人工介入。
*   **持久化**：消息写到磁盘后再回复 ack。
